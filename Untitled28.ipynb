```python
#Data Pre Processing
df.info()
```

```python
df.columns
```

```python
#Replacing all zero values with NaN 
df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']]= df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']].replace(0, np.NaN)
```

```python
df.head()
```

```python
#Count of NaN
df.isnull().sum()
```

```python
#Replascing all NaN values with Mean values
df['Pregnancies'].fillna(df['Pregnancies'].mean(), inplace= True)
df['Glucose'].fillna(df['Glucose'].mean(), inplace= True)
df['BloodPressure'].fillna(df['BloodPressure'].mean(), inplace= True)
df['SkinThickness'].fillna(df['SkinThickness'].mean(), inplace= True)
df['Insulin'].fillna(df['Insulin'].mean(), inplace= True)
df['BMI'].fillna(df['BMI'].mean(), inplace= True)
df['DiabetesPedigreeFunction'].fillna(df['DiabetesPedigreeFunction'].mean(), inplace= True)
```

```python
#Satistical Summary
df.describe()
```

```python
#Splitting training and testing data
X= df.iloc[:, :8]
X
```

```python
Y= df.iloc[:, -1]
Y
```

```python
#Splitting
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size= 0.2, random_state=2, stratify= df['Outcome'])
```

```python
X_train
```

```python
#feature Engineering
scaler= StandardScaler()
X_train= scaler.fit_transform(X_train)
X_train
```

```python
X_test= scaler.transform(X_test)
X_test
```

```python
#Training Model:
```

```python
#LOGISTIC REGRESSION
```

```python
log_reg=LogisticRegression()
log_reg.fit(X_train, Y_train)
```

```python
y_predict= log_reg.predict(X_train)
```

```python
accuracy_log= accuracy_score(Y_train, y_predict)
print(accuracy_log)
print(confusion_matrix(Y_train, y_predict))
print(print(classification_report(Y_train, y_predict)))
```

```python
# K-Nearest Neighbors Algorithm
```

```python
knn= KNeighborsClassifier()
knn.fit(X_train, Y_train)
y_predict= knn.predict(X_train)
```

```python
accuracy_knn= accuracy_score(Y_train, y_predict)
print(accuracy_knn)
print(confusion_matrix(Y_train, y_predict))
print(classification_report(Y_train, y_predict))
```

```python
#Support Vector Machine
```

```python
svc= SVC(kernel='linear', random_state= 42)
svc.fit(X_train, Y_train)
```

```python
SVC_predict= svc.predict(X_train)
```

```python
accuracy_svm= accuracy_score(Y_train, SVC_predict)
print(accuracy_svm)
print(confusion_matrix(Y_train, SVC_predict))
print(print(classification_report(Y_train, SVC_predict)))
```

```python
#Decision Tree Algorithm
```

```python
from sklearn.tree import DecisionTreeClassifier
dectree = DecisionTreeClassifier(criterion = 'entropy', random_state = 42)
dectree.fit(X_train, Y_train)
```

```python
dst_predict= dectree.predict(X_train)
accuracy_dst= accuracy_score(Y_train, dst_predict)
print(accuracy_dst)
print(confusion_matrix(Y_train, dst_predict))
print(print(classification_report(Y_train, dst_predict)))
```

```python
# Accuracy of each model 
```

```python
print("Logistic Regression: " +str((accuracy_log )*100))
print("Support Vector Classifier: " + str((accuracy_svm) * 100))
print("K Nearest neighbors: " + str(accuracy_knn * 100))
print("Decision tree: " + str(accuracy_dst* 100))
```
